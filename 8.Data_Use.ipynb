## 로그 파일 읽기

### 유효한 페이지 뷰 수 확인

```python
pageviews = 0
f = None
with open('data/log.txt', 'r') as f:
    logs = f.readlines()
    for log in logs:
        log = log.split()
        status = log[8]
        if status == '200':
            pageviews += 1
    print('총 페이지뷰: [%d]' %pageviews)
```

### IP 수 확인

```python
visit_ip = []
f = None
with open('data/log.txt', 'r') as f:
    logs = f.readlines()
    for log in logs:
        log = log.split()
        ip = log[0]
        if ip not in visit_ip:
            visit_ip.append(ip)
    print('방문자수: [%d]' %len(visit_ip))
```

### 트래픽 합계

```python
total_service = 0
with open('data/log.txt', 'r') as f:
    logs = f.readlines()
    for log in logs:
        log = log.split()
        servicebyte = log[9]
        if servicebyte.isdigit():
            total_service += int(servicebyte)
    total_service /= 1024
    print('총 서비스 용량: %dKB' %total_service)
```

### IP 별 트래픽 합계

```python
services = {}
f = None
with open('data/log.txt', 'r') as f:
    logs = f.readlines()
    for log in logs:
        log = log.split()
        ip = log[0]
        servicebyte = log[9]
        if servicebyte.isdigit():
            servicebyte = int(servicebyte)
        else:
            servicebyte = 0
        if ip not in services:
            services[ip] = servicebyte
        else:
            services[ip] += servicebyte
    ret = sorted(services.items(), key=lambda x: x[1], reverse=True)
    print('사용자IP – 서비스용량')
    for ip, b in ret:
        print('[%s] – [%d]' %(ip, b))
```

## 엑셀 파일 읽기 및 쓰기

```python
import openpyxl
wb = None
try:
    # 엑셀파일 열기
    wb = openpyxl.load_workbook('data/score.xlsx')
 
    # 현재 Active Sheet 얻기
    ws = wb.active
    # ws = wb.get_sheet_by_name("Sheet1“)

    for r in ws.rows:
        row_index = r[0].row   # 행 인덱스
        if row_index == 1:
            ws.cell(row=row_index, column=5).value = '합계'
            continue
        name = r[0].value
        kor = r[1].value
        eng = r[2].value
        math = r[3].value
        sum = kor + eng + math
        # 합계 쓰기
        ws.cell(row=row_index, column=5).value = sum
        print(name, ':' , kor, eng, math, sum)
    # 엑셀 파일 저장
    wb.save("data/score2.xlsx")
except Exception as e:
    print("예외 발생:", e)
finally:
    wb.close()
```

### 데이터 수정

```python
from pymongo import MongoClient 

# 데이터베이스 연결
conn = MongoClient('127.0.0.1')

# 데이터베이스 설정
db = conn.mymongo

# 컬렉션 설정
collect = db.users 

# 전체 문서 조회
result = collect.find()

# 조회 문서 출력
for r in result :
    print(r)
    
# 데이터 1개 수정    
collect.update_one(
    { "empno" : "10001" }, 
    { "$set" : 
        { "name" : "kkk" }
    }
)

# 전체 문서 조회
result = collect.find()

# 조회 문서 출력
for r in result :
    print(r)

#데이터 여러 개 수정
collect.update_many(
    { 'age':{'$gte':30} }, 
    { "$set" : 
        { "name" : "park" }
    }
)

# 전체 문서 조회
result = collect.find()

# 조회 문서 출력
for r in result :
    print(r)

# 데이터 삭제
collect.delete_many( {"age": { "$gte": 30 } } )

# 전체 문서 조회
result = collect.find()

# 조회 문서 출력
for r in result :
    print(r)
```

### 기상청 날씨 데이터 활용

```python
import requests                  # 웹 페이지의 HTML을 가져오는 모듈
from bs4 import BeautifulSoup    # HTML을 파싱하는 모듈

# 웹 페이지를 가져온 뒤 BeautifulSoup 객체로 만듦
response = requests.get('http://www.weather.go.kr/weather/observation/currentweather.jsp')

soup = BeautifulSoup(response.content, 'html.parser')

table = soup.select('.table_develop3')    # <table class="table_develop3">을 찾음

loc = []
temp =[]
hum=[]

# 데이터를 저장할 리스트 생성
for tr in table[0].find_all('tr'):      # 모든 <tr> 태그를 찾아서 반복(각 지점의 데이터를 가져옴)
    tds = list(tr.find_all('td'))    # 모든 <td> 태그를 찾아서 리스트로 만듦
    for td in tds:                   # <td> 태그 리스트 반복(각 날씨 값을 가져옴)
        if td.find('a'):             # <td> 안에 <a> 태그가 있으면(지점인지 확인)
            loc.append(td.find('a').text)    # <a> 태그 안에서 지점을 가져옴
            temp.append(tds[5].text)    # <td> 태그 리스트의 여섯 번째(인덱스 5)에서 기온을 가져옴
            hum.append(tds[9].text)       # <td> 태그 리스트의 열 번째(인덱스 9)에서 습도를 가져옴
print('도시 이름:', loc[0:5])
print('현재 온도:', temp[0:5])
print('습도:', hum[0:5])
print()
print()

cities= ['서울', '인천', '대전', '대구', '광주', '부산', '울산', '창원']
temperatures = []
humidities =[]
for city in cities:
    j = 0
    for c in loc:
        if city == c:
            temperatures.append(temp[j])
            humidities.append(hum[j])
            break
        j = j + 1
        
print('도시 이름:', cities)
print('현재 온도:', temperatures)
print('습도:', humidities)
```

### 동아일보 기사 검색

```python
#import datetime 
import requests, bs4
from io import StringIO
from html.parser import HTMLParser

#한글 URL을 처리하기 위한 모듈 import
from urllib.parse import quote


#문자열에서 태그를 제거하기 위한 클래스 와 함수
class MLStripper(HTMLParser):
    def __init__(self):
        super().__init__()
        self.reset()
        self.strict = False
        self.convert_charrefs= True
        self.text = StringIO()
    def handle_data(self, d):
        self.text.write(d)
    def get_data(self):
        return self.text.getvalue()
    
def strip_tags(html):
    s = MLStripper()
    s.feed(html)
    return s.get_data()


#오늘 날짜를 yyyymmdd 형식으로 만들기
#dt = datetime.datetime.now()
#print(dt)
#s = dt.strftime('%Y%m%d')
#print(s)

keyword = input("검색하고자 하는 내용:")
#print(keyword)
#입력받은 내용을 가지고 동아일보 기사 검색 url 만들기
openurl = 'http://www.donga.com/news/search?p=1&query=' + quote(keyword) + '&check_news=1&more=1&sorting=1&search_date=2&v1=&v2=&range=1'

print(openurl)

#html 다운로드 받기
links = requests.get(openurl)
#print(links.text)
       
#파싱 객체 생성
bs = bs4.BeautifulSoup(links.text, "html.parser")

#기사 건수를 가져오기 위한 파싱
article_cnt = bs.select("h2 > span")
#print(article_cnt)      
 
#첫번째 데이터를 가져와서 공백을 기준으로 분할
#print(article_cnt[0].getText())
sp = article_cnt[0].getText().split()
#print(sp)
cnt = int(sp[1])
#print(cnt)

#페이지 수 구하기
pagesu = int(cnt/15)+1
print(pagesu)

#기사 URL을 저장하기 위한 list 생성 
li = []
#페이지 별로 이동하면서 기사의 링크를 li에 추가하기
for idx in range(0,pagesu):
    openurl = 'https://www.donga.com/news/search?p=' + str(idx * 15 + 1) + '&query=' + quote(keyword) + '&check_news=1&more=1&sorting=1&search_date=2&v1=&v2=&range=1'
    links = requests.get(openurl)
    bs = bs4.BeautifulSoup(links.text, "html.parser")
    #실제 기사를 포함하고 있는 링크만 추출하기
    temp = bs.select("p.tit > a")
    for link in temp:
        addr = link.get("href")
        li.append(addr)
        
print(len(li)) #기사 링크 개수 확인
print(li) #기사 링크 출력


with  open("./data/" + keyword + ".txt", mode="a", encoding="utf8") as outputfile:
    for link in li:
        articleUrl = requests.get(link)
        articleUrl.encoding=None
        bs = bs4.BeautifulSoup(articleUrl.text, "html.parser")    
        print(articleUrl.text)
        temp = bs.select("#content > div > div.article_txt")
        print(temp)
        for link in temp:
            outputfile.write(strip_tags(link.getText()))
            
```
