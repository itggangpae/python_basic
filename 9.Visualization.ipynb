### 생일 기온 그래프

```python
import csv

#파일 열기
f = open('data/mokpo.csv', encoding='ms949')

#파일을 csv로 연결
data = csv.reader(f)
#첫번째 행의 데이터를 넘어가기
next(data)

#최고기온과 최저기온 및 년도를 저장하기 위한 list 생성
high = []
low = []
year = []

#데이터 저장하기
for row in data :
    #최고기온과 최저기온이 존재한다면
    if row[-1] != '' and row[-2] != '' :
        #첫번째 데이터를 .을 기준으로 분할
        date = row[0].split('.')
        #분할된 데이터의 첫번째가 8 이고 두번째 데이터가 18이라면
        if date[1] == '8' and date[2] == '18' :
            #최고기온, 최저기온, 년도를 list에 저장
            high.append(float(row[-1]))
            low.append(float(row[-2]))
            year.append(int(date[0]))
            
import platform
from matplotlib import font_manager, rc
import matplotlib

#그래프에 음수를 표현하기 위한 설정
matplotlib.rcParams['axes.unicode_minus'] = False 
#운영체제의 종류에 따른 폰트 설정
#매킨토시의 경우
if platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
#윈도우의 경우
elif platform.system() == 'Windows':
    font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
    rc('font', family=font_name)
    
import matplotlib.pyplot as plt
#그래프 영역 크기 설정
plt.figure(figsize=(10,4))

#그래프 설정
plt.plot(high, label='최고기온', lw=3, color='r', linestyle='-' , marker='s')
plt.plot(low, label='최저기온', lw=3, color='b', linestyle='-' , marker='s')

#그래프 옵션 설정
plt.ylim(-20, 40)
plt.yticks(range(-20,40,5))
plt.xticks(range(0,len(year),1), year, rotation='vertical')
plt.xlabel('년도')
plt.ylabel('최고 및 최저 기온')
plt.title('매년 내 생일의 기온')
plt.legend()

#그래프 출력
plt.show()
```

# 산포도

```python
#pop.csv 파일 읽기
import csv
import math
#파일의 내용을 읽어오기
f = open('data/pop.csv', encoding='ms949')
data = csv.reader(f)
#남자와 여자 의 인구를 저장할 list 생성
man = []
woman = []
#산점도에서 포인트의 크기를 위한 list
size = []

#동네 입력
location = input('성별 인구 분포를 확인할 지역을 입력하세요 : ')
for row in data :
    #첫번째 열에서 동네 이름이 포함된 경우에만
    if location in row[0] :
        for i in range(24,45) : 
            man.append(int(row[i].replace(',', '')))
            woman.append(int(row[i+23].replace(',', '')))
            #읽어온 데이터를 더해서 제곱근을 구한 후 size에 추가
            size.append(math.sqrt(int(row[i].replace(',', '')) + int(row[i+23].replace(',', '')))/5)
        break
print(man[0:5])
print(woman[0:5])
print(size[0:5])
```

```python
import matplotlib.pyplot as plt
import platform
from matplotlib import font_manager, rc
#매킨토시의 경우
if platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
#윈도우의 경우
elif platform.system() == 'Windows':
    font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
    rc('font', family=font_name)

#그래프 그릴 영역 만들기
plt.figure(figsize = (4,2))   
#그래프 그리기
plt.scatter(x = man, y=woman, marker='1', s = size, c = range(21), alpha=0.5, cmap='jet')
plt.colorbar()
plt.title(location + ' 성별 인구 분포')
plt.xlabel('남성 인구 수')
plt.ylabel('여성 인구 수')

#plt.show 앞에 추가
plt.plot(range(max(man)),range(max(man)), 'g')

#그래프 출력
plt.show()
```

# 대중교통 데이터 분석

## 유임승차 비율이 가장 높은 10개의 데이터를 막대그래프로 출력하기

```python
import openpyxl
import xlrd

import matplotlib.pyplot as plt
import platform
from matplotlib import font_manager, rc
#매킨토시의 경우
if platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
#윈도우의 경우
elif platform.system() == 'Windows':
    font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
    rc('font', family=font_name)
    
wb = None
data = []

#데이터 정렬을 위한 함수
def ratio(temp):
    if temp[2] != 0:
        return temp[1]/temp[2]
    else:
        return 0;

try:
    # 엑셀파일 열기
    wb = xlrd.open_workbook('data/traffic.xls')
    #2번째 Sheet 얻기
    ws = wb.sheet_by_index(2)
    #모든 행을 순회하면서
    for row in ws.get_rows():
        rowdata = []
        #각 셀의 데이터를 행 단위로 모아서
        for cell in row:
            rowdata.append(cell.value)
        #list에 저장 - 호선이름_역이름, 유임승차인원, 무임승차인원
        data.append((rowdata[1]+' ' + rowdata[3], rowdata[4], rowdata[6]))
    #가져온 데이터 확인
    #print(data)
    
    #무임승차 인원과 유임승차 인원을 숫자 데이터로 변경
    #,제거
    tempdata = []
    for imsi in data:
        tempdata.append((imsi[0], imsi[1].replace(',', ''),imsi[2].replace(',', '')))
    #첫번째 데이터 제거
    del tempdata[0]
    
    #2번째 데이터와 3번째 데이터를 숫자로 변환해서 저장
    data = []
    for imsi in tempdata:
        data.append((imsi[0], int(imsi[1]),int(imsi[2])))
    #print(data)
    
    data.sort(key=ratio, reverse=True)
    #print(data)
    
    graphdata = []
    labels = []
    for i in range(0,10):
        graphdata.append(ratio(data[i]))
        labels.append(data[i][0])
    
    #그래프 그릴 영역 만들기
    plt.figure(figsize = (4,2), dpi=150)   
    plt.bar(range(0,10,1), graphdata , width=0.7, color='r')
    plt.xticks(range(0,10,1),labels, rotation='vertical')
    #그래프 출력
    plt.show()
    
except Exception as e:
    print("예외 발생:", e)
```

## 모든 승하차 인원이 가장 많은 10개역의 승하차 인원의 비율을 파이그래프로 그리기

```python
import openpyxl
import xlrd

import matplotlib.pyplot as plt
import platform
from matplotlib import font_manager, rc
#매킨토시의 경우
if platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
#윈도우의 경우
elif platform.system() == 'Windows':
    font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
    rc('font', family=font_name)
plt.rcParams.update({'font.size': 6})    
    
wb = None
data = []

#데이터 정렬을 위한 함수
def total(temp):
    return temp[1] + temp[2] + temp[3] + temp[4]

try:
    # 엑셀파일 열기
    wb = xlrd.open_workbook('data/traffic.xls')
    #2번째 Sheet 얻기
    ws = wb.sheet_by_index(2)
    #모든 행을 순회하면서
    for row in ws.get_rows():
        rowdata = []
        #각 셀의 데이터를 행 단위로 모아서
        for cell in row:
            rowdata.append(cell.value)
        #list에 저장 - 호선이름_역이름, 유임승차인원, 무임승차인원
        data.append((rowdata[1]+' ' + rowdata[3], rowdata[4], rowdata[5], rowdata[6], rowdata[7]))
    #가져온 데이터 확인
    #print(data)
    
    #무임승차 인원과 유임승차 인원을 숫자 데이터로 변경
    #,제거
    tempdata = []
    for imsi in data:
        #지하철 역 이름 및 유임승하차, 무임승하차 인원을 가져오기
        tempdata.append((imsi[0], imsi[1].replace(',', ''),imsi[2].replace(',', ''), imsi[3].replace(',', ''),imsi[4].replace(',', '')))
    #첫번째 데이터 제거
    del tempdata[0]
    data = []
    i = 0
    #유임승하자, 무임승하차 인원을 정수 데이터로 변환
    for imsi in tempdata:
        data.append((imsi[0],int(imsi[1]), int(imsi[2]), int(imsi[3]), int(imsi[4])))
    #print(data)
    
    #승하차 인원의 합으로 데이터 정렬
    data.sort(key=total, reverse=True)
    #print(data)
    
    
    graphdata = []
    labels = ['유임승차', '유임하차', '무임승차', '무임하차']
    subways = []
    for i in range(0,10):
        graphdata.append([data[i][1], data[i][2], data[i][3], data[i][4]])
        subways.append(data[i][0])
    #print(graphdata)
    
    
    #그래프 그릴 영역 만들기
    fig = plt.figure(figsize = (8,3), dpi=200)
    ax = []
    for i in range(1,11,1):
        ax.append(fig.add_subplot(2,5,i))

    for i in range(0,10,1):
        ax[i].pie(graphdata[i], labels=labels, autopct='%1.1f%%', startangle=90)
        ax[i].title.set_text(subways[i])
        ax[i].axis('equal')
    #그래프 출력
    plt.show()
    
except Exception as e:
    print("예외 발생:", e)
```

##  시간대 별 승하차 인원 추이 파악 

```python
import openpyxl
import xlrd

import matplotlib.pyplot as plt
import platform
from matplotlib import font_manager, rc
#매킨토시의 경우
if platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
#윈도우의 경우
elif platform.system() == 'Windows':
    font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
    rc('font', family=font_name)
plt.rcParams.update({'font.size': 6})    
    
data = []

def comma(temp):
    return temp.replace(',', '')

station = input('승하차 인원 추이를 알고자 하는 역이름:')

try:
    # 엑셀파일 열기
    wb = xlrd.open_workbook('data/traffic.xls')
    #3번째 Sheet 얻기
    ws = wb.sheet_by_index(3)
    #모든 행을 순회하면서
    i = 0
    for row in ws.get_rows():
        i = i + 1
        if i==1 or i==2:
            continue
        
        rowdata = []
        #각 셀의 데이터를 행 단위로 모아서
        for cell in row:
            rowdata.append(cell.value)
        del rowdata[2]
        del rowdata[1]
        del rowdata[0]
        del rowdata[-1]
        
        rowdata = list(map(comma, rowdata))
        rowdata[1:] = list(map(int, rowdata[1:]))
        if rowdata[0] == station:
            del rowdata[0]
            data.append(rowdata)
    #print(data)
    
    stationdata = [0] * 48
    
    for row in data:
        for i in range(0,48,1):
            stationdata[i] = stationdata[i] + row[i]
    #print(stationdata)
    
    hour = []
    for i in range(0,24,1):
        hour.append((i+4)%24)
    
    plt.figure(figsize = (4,2), dpi=200)
    plt.title(station + '역 시간대별 승하차 인원 추이')
    plt.plot(stationdata[0:48:2], label = '승차')
    plt.plot(stationdata[1:48:2], label = '하차')
    plt.xticks(range(0,24,1),hour)
    plt.legend()
    
    plt.show()
    
except Exception as e:
    print("예외 발생:", e)
```

```python
import openpyxl
import xlrd

import matplotlib.pyplot as plt
import platform
from matplotlib import font_manager, rc
#매킨토시의 경우
if platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
#윈도우의 경우
elif platform.system() == 'Windows':
    font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
    rc('font', family=font_name)
plt.rcParams.update({'font.size': 6})    
    
data = []

def comma(temp):
    return temp.replace(',', '')
       
try:
    # 엑셀파일 열기
    wb = xlrd.open_workbook('data/traffic.xls')
    #3번째 Sheet 얻기
    ws = wb.sheet_by_index(3)
    #모든 행을 순회하면서
    i = 0
    for row in ws.get_rows():
        i = i + 1
        if i==1 or i==2:
            continue
        
        rowdata = []
        #각 셀의 데이터를 행 단위로 모아서
        for cell in row:
            rowdata.append(cell.value)
        del rowdata[2]
        del rowdata[0]
        del rowdata[-1]
        
        rowdata = list(map(comma, rowdata))
        rowdata[2:] = list(map(int, rowdata[2:]))
        
        data.append(rowdata)
    #가져온 데이터 확인
    #print(data)
    
    maxdata = []
    maxstation = []
    maxvalue = [0] * 24
    
    for j in range(0,24,1):
        temp = ()
        for row in data:
            if maxvalue[j] < row[(j+1)*2] + row[(j+1)*2+1]:
                maxvalue[j] = row[(j+1)*2] + row[(j+1)*2+1]
                tempstation = row[0] + ' ' + row[1]
        maxdata.append(maxvalue[j])
        maxstation.append(tempstation + '(' + str((j+4)%24) + ')')
    
    plt.figure(figsize = (4,2), dpi=200)
    plt.bar(range(0,24,1), maxdata, color = 'b')
    plt.xticks(range(0,24,1),maxstation, rotation = 90)
    plt.show()
    
except Exception as e:
    print("예외 발생:", e)
```
